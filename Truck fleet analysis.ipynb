```python
#data split
from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train, y_test = train_test_split(regdata.drop(['driverid','event_ind','city'], axis=1),
                                                    regdata['event_ind'],
                                                    test_size=0.3,
                                                    random_state=0)

X_train1.shape, X_test1.shape
```

```python
pip install feature_engine
```

```python
from sklearn.pipeline import Pipeline

from feature_engine.variable_transformers import PowerTransformer
from feature_engine.discretisers import EqualFrequencyDiscretiser
from feature_engine.categorical_encoders import OneHotCategoricalEncoder
```

```python
# make list of variables  types

# numerical: discrete vs continuous
discrete = [var for var in X_train1.columns if X_train1[var].dtype!='O' and var!='riskfactor' and X_train1[var].nunique()<10]
continuous = [var for var in X_train1.columns if X_train1[var].dtype!='O' and var!='riskfactor' and var not in discrete]


# categorical
categorical = [var for var in X_train1.columns if X_train1[var].dtype=='O']

print(discrete)
print(continuous)
print(categorical)
```

```python
#Preprocessing
model_pipe= Pipeline([
    ('rec', PowerTransformer(variables=['velocity', 'Events',
                                             'Totmiles'])),
    ('desc', EqualFrequencyDiscretiser(q=50,
                                       variables=continuous)),
    ('categorical_encoder', OneHotCategoricalEncoder(drop_last=False,
                                          variables=categorical))
])
```

```python
model_pipe.fit(X_train1, y_train)
```

```python
X_train=model_pipe.transform(X_train1)
X_test=model_pipe.transform(X_test1)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

clf = LogisticRegression(max_iter=1000).fit(X_train, y_train)

cv_scores = cross_val_score(clf, X_train, y_train)
train_preds=clf.predict(X_train)
test_preds= clf.predict(X_test)
# Mean Cross validation Score
print("Mean Cross-validation scores: {}".format(cv_scores.mean()))
print()

# Print Co-efficients
print("Logistic.coef_:", clf.coef_)
print("Logistic.intercept_:", clf.intercept_)

# Check test data set performance
print("Logistic Performance: ", clf.score(X_test,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {   
    'class_weight': [{0:100,1:1}, {0:10,1:1}, {0:1,1:1}, {0:1,1:10}, {0:1,1:100},'balanced'],
}
cgrid_logreg= GridSearchCV(LogisticRegression(solver='newton-cg'), param_grid, cv=5, n_jobs=2, scoring='f1')
cgrid_logreg.fit(X_train, y_train)
print("Best parameters: {}".format(cgrid_logreg.best_params_))
print("Best Mean cross-validation score: {:.2f}".format(cgrid_logreg.best_score_))
print("Logistic Performance: ", cgrid_logreg.score(X_test,y_test))
```

```python
# Print Co-efficients
print("Logistic.coef_:", cgrid_logreg.best_estimator_.coef_)
print("Logistic.intercept_:", cgrid_logreg.best_estimator_.intercept_)
```
